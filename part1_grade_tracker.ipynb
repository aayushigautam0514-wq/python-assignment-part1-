```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    name = student["name"].strip().title()
    roll = int(student["roll"])
    marks_list = student["marks_str"].split(", ")
    marks = []
    for m in marks_list:
        marks.append(int(m))

    cleaned = {"name": name, "roll": roll, "marks": marks}
    cleaned_students.append(cleaned)

    is_valid = True
    for word in name.split():
        if not word.isalpha():
            is_valid = False
            break

    if is_valid:
        validity = "✓ Valid name"
    else:
        validity = "✗ Invalid name"

    print("================================")
    print(f"Student : {name}   {validity}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("================================")
    print()

for student in cleaned_students:
    if student["roll"] == 103:
        print("ALL CAPS  :", student["name"].upper())
        print("lowercase :", student["name"].lower())

```

```python
##task-02

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]


print("Student:", student_name)
print()
print("Subject      Marks  Grade")
print("-" * 28)

for i in range(len(subjects)):
    mark = marks[i]
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
    print(f"{subjects[i]:<12} {mark:>5}  {grade:>5}")

total = 0
for mark in marks:
    total = total + mark

average = round(total / len(marks), 2)

highest_mark = marks[0]
highest_index = 0
lowest_mark = marks[0]
lowest_index = 0

for i in range(len(marks)):
    if marks[i] > highest_mark:
        highest_mark = marks[i]
        highest_index = i
    if marks[i] < lowest_mark:
        lowest_mark = marks[i]
        lowest_index = i

print()
print("Total Marks   :", total)
print("Average Marks :", average)
print("Highest       :", subjects[highest_index], "(", highest_mark, ")")
print("Lowest        :", subjects[lowest_index], "(", lowest_mark, ")")


print()
print("-- Add New Subjects (type 'done' to stop) --")

new_count = 0

while True:
    subject_input = input("Enter subject name: ").strip()

    if subject_input.lower() == "done":
        break

    if subject_input == "":
        print("Subject name cannot be empty. Try again.")
        continue

    mark_input = input("Enter marks for " + subject_input + " (0-100): ").strip()

    is_numeric = True
    for ch in mark_input.replace(".", "", 1):
        if not ch.isdigit():
            is_numeric = False
            break

    if not is_numeric or mark_input == "":
        print("Invalid input - marks must be a number. Entry skipped.")
        continue

    new_mark = float(mark_input)

    if new_mark < 0 or new_mark > 100:
        print("Invalid input - marks must be between 0 and 100. Entry skipped.")
        continue

    subjects.append(subject_input.title())
    marks.append(int(new_mark))
    new_count = new_count + 1
    print(subject_input.title(), "added successfully.")
    print()

updated_total = 0
for mark in marks:
    updated_total = updated_total + mark

updated_average = round(updated_total / len(marks), 2)

print()
print("New subjects added :", new_count)
print("Updated average    :", updated_average)
print("All subjects now   :", subjects)
```

```python
##task-03

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

results = []

print(f"{'Name':<18}| {'Average':^7} | {'Status'}")
print("-" * 40)

for i in range(len(class_data)):
    name = class_data[i][0]
    marks = class_data[i][1]

    total = 0
    for mark in marks:
        total = total + mark

    average = round(total / len(marks), 2)

    if average >= 60:
        status = "Pass"
    else:
        status = "Fail"

    results.append((name, average, status))
    print(f"{name:<18}| {average:^7.2f} | {status}")

passed = 0
failed = 0
topper_name = results[0][0]
topper_avg  = results[0][1]
total_avg   = 0

for i in range(len(results)):
    total_avg = total_avg + results[i][1]
    if results[i][2] == "Pass":
        passed = passed + 1
    else:
        failed = failed + 1
    if results[i][1] > topper_avg:
        topper_avg  = results[i][1]
        topper_name = results[i][0]

class_average = round(total_avg / len(results), 2)

print()
print("Students Passed :", passed)
print("Students Failed :", failed)
print("Class Topper    :", topper_name, "(", topper_avg, ")")
print("Class Average   :", class_average)

```
